## Cubic Capacity Validation

```python
import pandas as pd

# Load the Excel file into a DataFrame
df = pd.read_excel(r'C:\Users\ian.muthengi\Downloads\MERGED MOTOR 201223.xlsx')

#A function that validates cc column based on the value being null, less than 6 or greater than 10000
def validate_cubic_capacity(row):
    if pd.isna(row['CUBIC_CAPACITY']) or row['CUBIC_CAPACITY'] < 6 or row['CUBIC_CAPACITY'] > 10000:
        return 'INVALID'
    else:
        return 'VALID'
    
# Create a new column 'CUBIC_CAPACITY_VALIDATION' and append it to the dataframe
df['CUBIC_CAPACITY_VALIDATION'] = df.apply(validate_cubic_capacity, axis=1)

# Count the number of Valid and Invalid records
validation_counts = df['CUBIC_CAPACITY_VALIDATION'].value_counts()

print(validation_counts)
```

## Carry Capacity Validation

```python
# A function that validates carry capacity column based on the value being null, 1 or greater than 10
def validate_carry_capacity(row):
    if pd.isna(row['CARRY_CAPACITY']) or row['CARRY_CAPACITY'] ==1 or row['CARRY_CAPACITY'] > 10:
        return 'INVALID'
    else:
        return 'VALID'

# Create a new column 'CARRY_CAPACITY_VALIDATION' based on above conditions and append it to the dataframe
df['CARRY_CAPACITY_VALIDATION'] = df.apply(validate_carry_capacity, axis=1)

# Count the number of Valid and Invalid records
validation_counts = df['CARRY_CAPACITY_VALIDATION'].value_counts()

print(validation_counts)
```
